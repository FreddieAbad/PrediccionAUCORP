# Lectura Excel y Conversion a CSV
#### El siguiente script permite transformar un archivo Excel tipo xlsx (Formato >2010) en un archivo csv con las fechas seteadas de tal forma que se pueda realizar la lectura de los datos en los demas archivos donde se modela redes neuronales.

```python
import pandas as pd
def listoflists2list(listoflist): 
    for i in listoflist: 
        if type(i) == list: 
            listoflists2list(i) 
        else: 
            valoresPredecir.append(i) 
            
df = pd.read_excel (r'todas_las_variables.xlsx', encoding='utf-8')
#Lectura y transformacion de la celda de meses donde enero es 01-01 (1 de enero), febrero es 02-01(1 de febrero), etc.
df1=df.iloc[:,0:1]
df1=df1.apply(lambda x:x.replace("enero","01-01").replace("febrero", "02-01").replace("marzo", "03-01").replace("abril","04-01").replace("mayo", "05-01").replace("junio", "06-01").replace("julio","07-01").replace("agosto", "08-01").replace("septiembre", "09-01").replace("octubre","10-01").replace("noviembre", "11-01").replace("diciembre", "12-01"))
#Lectura de la celda de Ano
dfn=df.iloc[:,1:2] 
#Transformacion a Listas
dfnn=dfn.values.tolist()
df11=df1.values.tolist()
#Tratamiento de datos 
valoresPredecir=[]
listoflists2list(dfnn)
listaa=valoresPredecir
valoresPredecir=[]
listoflists2list(df11)
listab=valoresPredecir
#Union a un solo tipo de celda
i=0
bandera=True
nuevalista=[]
while bandera:
    nuevalista.append(str(listaa[i])+'-'+listab[i])
    i+=1
    if(i<len(listaa)):
        bandera=True
    else:
        bandera=False 
#Insercion en la demas data 
df.insert(loc=0, column='Fecha', value=nuevalista)
cat=['A' for i in range(len(dfnn))]
df.insert(loc=1,column='CATEGORIA',value=cat)
#Eliminacion de datos de mes y ano, innecesarios en este punto
df = df.drop('mes de analisis', 1)
df = df.drop('A帽o', 1)
print(df)
df.to_csv('aucorpTrainActualizado.csv',index=False)
```
