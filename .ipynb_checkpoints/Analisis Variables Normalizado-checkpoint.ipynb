```python
import numpy
import matplotlib.pyplot as plt
from pandas import read_csv
import math
from matplotlib.pyplot import figure
from pandas import DataFrame, concat, read_csv
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
dataset = read_csv('aucorpTrain.csv', header=0, index_col=0)
dataset=dataset.iloc[:,2:]
#print(dataset)
values = dataset.values
encoder = LabelEncoder()
values[:,4] = encoder.fit_transform(values[:,4])
#print(values)
values = values.astype('float32')
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
scaled = scaler.fit_transform(values)
#print(scaled[:,2]) 
```

```python
def graficaAtributo(atributo,nombreLeyenda,titulo,xetiqueta,yetiqueta,nombreArchivo,alt=6,ancho=16):
    figure(figsize=(ancho,alt))
    plt.plot(scaled[:,atributo], "-o")
    plt.legend([nombreLeyenda], loc = 'upper right')
    plt.grid(True)
    plt.title(titulo, fontsize = 15)
    plt.xlabel(xetiqueta, fontsize = 10)
    plt.ylabel(yetiqueta, fontsize = 10)
    plt.xticks(fontsize=7, color = 'g', rotation = 90) 
    #plt.savefig(nombreArchivo, dpi=300)
```

```python
#Grafica del atributo a predecir
graficaAtributo(0,'Ventas',"Ventas 2012-2018","Tiempo (Mes&Ao)","Ventas",'ventasGrafica.png')
```
